```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,100)
y = x *2
z = x **2
print('Hi I am Asish')
plt.show()
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax1 = fig.add_axes([0.1,0.5,0.3,0.3])
ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')
ax.set_title('X vs Z')
ax1.plot(x,y)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('X vs Y')
ax1.set_xlim([20, 30])
ax1.set_ylim([40, 50])
```
