# Matplot Py Lib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y =  x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.xlabel('X label')
plt.ylabel('Y label')
plt.plot(x,y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(x,y,'b')
```

Plotting using axes

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('X Vs Y')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.9,0.9])
axes2 = fig.add_axes([0.2,0.5,0.2,0.2])

axes1.plot(x,y)
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('X Vs Y')

axes2.plot(y,x)
axes2.set_xlabel('X Label')
axes2.set_ylabel('Y Label')
axes2.set_title('y Vs x')
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
axes
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y)
axes[0].set_title('First_Plot')

axes[1].plot(x,y)
axes[1].set_title('Second_Plot')
```

# Figure Size and DPI

```python
fig = plt.figure(figsize = (8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.jpg')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label = 'X Squared')
ax.plot(x,x**3, label = 'X Cubed')
ax.legend(loc = 2)
```

# Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'purple', linewidth = 2, linestyle = '--',marker = 'o', markersize = 20, markerfacecolor = 'yellow', markeredgewidth = 3 , markeredgecolor= 'green')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
